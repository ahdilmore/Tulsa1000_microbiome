```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import biom
```

```python
feature_key = pd.read_csv('/projects/u19/Tulsa/Medications_Data/feature_key.csv', index_col=0)
gg2_taxonomy = pd.read_csv('gg2_taxonomy_16s.csv')
taxonomy = feature_key.reset_index().merge(gg2_taxonomy, left_on='sequences', right_on='Feature ID')
taxonomy[['domain', 'phylum', 'class', 'order', 'family', 'genus', 'species']] = taxonomy['Taxon'].str.split(';', expand=True)
taxonomy['species'] = [s[4:] if len(s) > 4 else 'unknown' for s in taxonomy['species']]

species_key = dict(zip(taxonomy['index'], taxonomy['species']))

def read_results(p):
    return pd.read_csv(p, sep='\t', index_col='Feature', dtype={'Feature': 'str'})
def unpack_hdi_and_filter(df, col, lower, upper):
    df[['lower', 'upper']] = df[col].str.split(',', expand=True)
    # remove ( from lower and ) from upper and convert to float 
    df['lower'] = df['lower'].str[1:].astype('float')
    df['upper'] = df['upper'].str[:-1].astype('float')
    df['significant'] = np.where((df.lower > 0) | (df.upper < 0), 'yes', 'no')
    df['direction'] = np.where((df['significant'] == 'yes') & (df['lower'] < 0), lower, 
                                np.where((df['significant'] == 'yes') & (df['upper'] > 0), upper, np.nan))
    df['lower'] = df[col.replace('hdi', 'mean')] - df['lower'] 
    df['upper'] = df['upper'] - df[col.replace('hdi', 'mean')]
    only_sigs = df.loc[df['significant'] == 'yes']
    only_sigs['taxonomy'] = [species_key[s] if s in species_key else 'unknown' for s in only_sigs.index]
    return only_sigs
```

```python
key = {
    'Group_Anxiety': ['anxiety.beta_var.tsv', 'DX_ANX[T.True]_', 'HC', 'ANX'], 
    'Group_Depression': ['depression.beta_var.tsv', 'DX_MDD[T.True]_', 'HC', 'MDD'], 
    'Medications_Anxiety': ['anx_table.beta_var.tsv', 'N05B_Anxiolytics[T.True]_', 'Unmedicated ANX', 'Anxiolytics'],
    'Medications_Depression': ['mdd_table.beta_var.tsv', 'N06A_Antidepressants[T.True]_', 'Unmedicated MDD', 'Antidepressants'], 
}

all_results = {}
```

```python
for k in key: 
    all_results[k] = unpack_hdi_and_filter(read_results(key[k][0]), key[k][1]+'hdi', key[k][2], key[k][3])
```

```python
anx_enriched = all_results['Group_Anxiety'].loc[all_results['Group_Anxiety']['direction'] == 'ANX']
anx_hc = all_results['Group_Anxiety'].loc[all_results['Group_Anxiety']['direction'] == 'HC']
mdd_enriched = all_results['Group_Depression'].loc[all_results['Group_Depression']['direction'] == 'MDD']
mdd_hc = all_results['Group_Depression'].loc[all_results['Group_Depression']['direction'] == 'HC']
```

```python
mdd_and_anx = mdd_enriched.index.intersection(anx_enriched.index)
hc = anx_hc.index.intersection(mdd_hc.index)
```

```python
for m in mdd_and_anx: 
    if species_key[m] != 'unknown': 
        print(species_key[m])
```

```python
for m in hc: 
    if species_key[m] != 'unknown': 
        print(species_key[m])
```

```python
anxiolytics_enriched = all_results['Medications_Anxiety'].loc[all_results['Medications_Anxiety']['direction'] == 'Anxiolytics']
no_med_anx = all_results['Medications_Anxiety'].loc[all_results['Medications_Anxiety']['direction'] == 'Unmedicated ANX']
antidepressants_enriched = all_results['Medications_Depression'].loc[all_results['Medications_Depression']['direction'] == 'Antidepressants']
no_med_mdd = all_results['Medications_Depression'].loc[all_results['Medications_Depression']['direction'] == 'Unmedicated MDD']
```

```python
med_enriched = anxiolytics_enriched.index.intersection(antidepressants_enriched.index)
no_med = no_med_anx.index.intersection(no_med_mdd.index)
```

```python
for m in med_enriched: 
    if species_key[m] != 'unknown': 
        print(species_key[m])
```

```python
for m in no_med: 
    if species_key[m] != 'unknown': 
        print(species_key[m])
```

```python
anx = biom.load_table('/projects/u19/Tulsa/Medications_Data/anx_table.biom').to_dataframe().T
mdd = biom.load_table('/projects/u19/Tulsa/Medications_Data/mdd_table.biom').to_dataframe().T
unmed = biom.load_table('/projects/u19/Tulsa/Medications_Data/unmedicated_table.biom').to_dataframe().T
tulsa_table = pd.concat([anx, mdd, unmed]).drop_duplicates()
```

```python
anx_md = pd.read_csv('/projects/u19/Tulsa/Medications_Data/anx_metadata.tsv', sep='\t', index_col=0)
mdd_md = pd.read_csv('/projects/u19/Tulsa/Medications_Data/mdd_metadata.tsv', sep='\t', index_col=0) 
unmed_md = pd.read_csv('/projects/u19/Tulsa/Medications_Data/unmedicated_metadata.tsv', sep='\t', index_col=0)
tulsa_md = pd.concat([anx_md, mdd_md, unmed_md]).drop_duplicates()
```

```python
tulsa_md['category'] = np.where((tulsa_md['GroupAssignment'] == 'Mood/Anxiety') & ((tulsa_md['N06A_Antidepressants'] == True) | (tulsa_md['N05B_Anxiolytics'] == True)), 
                                'Medicated Mood/Anxiety', np.where(tulsa_md['GroupAssignment'] == 'Healthy Control', 'Unmedicated HC', 'Unmedicated Mood/Anxiety'))
tulsa_md['mdd_anx_ratio'] = np.log((tulsa_table[mdd_and_anx].sum(axis=1)+1) / (tulsa_table[hc].sum(axis=1)+1))
tulsa_md['medications_ratio'] = np.log((tulsa_table[med_enriched].sum(axis=1)+1) / (tulsa_table[no_med].sum(axis=1)+1))
```

```python
unmed_md['mdd_anx_ratio'] = np.log((unmed[mdd_and_anx].sum(axis=1)+1) / (unmed[hc].sum(axis=1)+1))
mdd_md['mdd_medicated'] = np.where((mdd_md['DX_MDD'] == True) & (mdd_md['N06A_Antidepressants']), True, False)
#mdd_md['mdd_anx_ratio'] = np.log((mdd[mdd_and_anx].sum(axis=1)+1) / (mdd[hc].sum(axis=1)+1))
```

```python
tulsa_md['mdd_anx_ratio'] = np.log((tulsa_table[mdd_and_anx].sum(axis=1)+1) / (tulsa_table[hc].sum(axis=1)+1))
```
